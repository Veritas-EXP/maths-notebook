## Part 2: Python Code Cells

Code cells run Python code. Run the next cell with `Shift + Enter`.

```python
name = "Student"
age = 12

print("Hello,", name)
print("Age:", age)
```

## Part 3: Using `print()`

`print()` shows values and messages in output.

```python
a = 8
b = 3

print("a =", a)
print("b =", b)
print("a + b =", a + b)
print("a * b =", a * b)
```

## Part 4: Notebook State

Variables from earlier cells can be used in later cells.

Run the next cell after running the previous code cells.

```python
total = a + b
print("total =", total)
```

## Part 5: Modules and `import`

A module is a file with useful code written by others.

- `import math` gives math tools
- `import matplotlib.pyplot as plt` gives plotting tools
- `as plt` creates a short nickname

```python
import math
import matplotlib.pyplot as plt

print("sqrt(16) =", math.sqrt(16))
print("matplotlib loaded as plt")
```

## Part 6: First Tiny Plot

Now use `matplotlib` to draw a simple line.

```python
x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]

plt.plot(x, y, marker="o", label="y = x^2")
plt.title("My First Plot")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()
```
