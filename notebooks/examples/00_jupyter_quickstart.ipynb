- You can write text notes.

## 2) Simple calculator
Try changing the numbers, then run again.

```python
a = 12
b = 5
print("a + b =", a + b)
print("a - b =", a - b)
print("a * b =", a * b)
print("a / b =", a / b)
```

## 3) Variables

## 5) One tiny NumPy example

```python
import numpy as np
numbers = np.array([2, 4, 6, 8])
numbers.mean()
```

## 6) One tiny SymPy example (algebra)

```python
import sympy as sp
x = sp.symbols("x")
sp.factor(x**2 + 5*x + 6)
```

## 7) One tiny graph (Matplotlib)

```python
import matplotlib.pyplot as plt
x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]
plt.plot(x, y, marker="o")
plt.title("My First Graph")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

## 8) Cell types and useful shortcuts
- Code cell: press `Y` (in command mode)
- Markdown cell: press `M` (in command mode)
- Run cell: `Shift+Enter`
- Edit mode: `Enter`
- Command mode: `Esc`
